```python
import pandas as pd
df_results = pd.read_csv('Fillout Acadoscope results.csv')
df_planner = pd.read_csv('Acadoscope_Survey Planner - acadoscope nambah dikit.csv')
df_results.info()
```

```python
kolom_sampah = ['Submission ID', 'Last updated', 'Submission started', 'Status', 
                'Current step', 'Created records', 'Errors', 'Url', 'Network ID', 'Nama lengkap', 'Instagram']
df_clean = df_results.drop(columns=kolom_sampah)
nama_kolom_baru = [
    'NIM', 'Kelas', 'Adaptasi', 'Kondisi_Akademik', 'Hambatan', 
    'Penyebab_Kurang_Maksimal', 'Metode_Belajar', 'Minat_Bantuan', 
    'Program_Dibutuhkan', 'Matkul_Sulit', 'Waktu_Ideal', 
    'Kehadiran_Akpres', 'Bantuan_Lomba', 'Harapan_Ormawa'
]
df_clean.columns = nama_kolom_baru
df_clean = df_clean.fillna('Tidak ada jawaban')
display(df_clean.head())
df_clean.info()
```

```python
df_clean['Kode_Fakultas'] = df_clean['NIM'].str[0].str.upper()
distribusi_fakultas = df_clean['Kode_Fakultas'].value_counts()
print("Distribusi Responden per Fakultas:")
print(distribusi_fakultas)
distribusi_kelas = df_clean['Kelas'].value_counts().head(10)
print("\nTop 10 Kelas dengan Responden Terbanyak:")
print(distribusi_kelas)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
distribusi_fakultas_clean = df_clean['Kode_Fakultas'].value_counts().reset_index()
distribusi_fakultas_clean.columns = ['Fakultas', 'Jumlah']
distribusi_kelas_clean = df_clean['Kelas'].value_counts().reset_index()
distribusi_kelas_clean.columns = ['Kelas', 'Jumlah']
plt.figure(figsize=(12, 6))
ax1 = sns.barplot(data=distribusi_fakultas_clean, x='Fakultas', y='Jumlah', palette='viridis')
plt.title('Distribusi Responden per Fakultas', fontsize=14, fontweight='bold')
plt.xlabel('Kode Fakultas', fontsize=12)
plt.ylabel('Jumlah Responden', fontsize=12)
ax1.bar_label(ax1.containers[0]) 
plt.show()
plt.figure(figsize=(10, 16))
ax2 = sns.barplot(data=distribusi_kelas_clean, x='Jumlah', y='Kelas', palette='magma')
plt.title('Distribusi Responden per Kelas (Seluruh Data)', fontsize=14, fontweight='bold')
plt.xlabel('Jumlah Responden', fontsize=12)
plt.ylabel('Nama Kelas', fontsize=12)
ax2.bar_label(ax2.containers[0], padding=3)
plt.tight_layout()
plt.show()
```

```python
print("REKAP JUMLAH RESPONDEN PER FAKULTAS")
for index, row in distribusi_fakultas_clean.iterrows():
    print(f"Fakultas {row['Fakultas']} : {row['Jumlah']} mahasiswa")
print("\nREKAP JUMLAH RESPONDEN PER KELAS")
for index, row in distribusi_kelas_clean.iterrows():
    # Menggunakan ljust(7) biar spasinya rapi sejajar
    print(f"Kelas {row['Kelas'].ljust(7)} : {row['Jumlah']} mahasiswa")
```

```python
distribusi_adaptasi = df_clean['Adaptasi'].value_counts()
distribusi_kondisi = df_clean['Kondisi_Akademik'].value_counts()
hambatan_split = df_clean['Hambatan'].str.split(',').explode().str.strip()
distribusi_hambatan = hambatan_split.value_counts().reset_index()
distribusi_hambatan.columns = ['Jenis Hambatan', 'Jumlah']
print("TOP 5 HAMBATAN AKADEMIK MAHASISWA PKU")
for index, row in distribusi_hambatan.head(5).iterrows():
    print(f"- {row['Jenis Hambatan']} : {row['Jumlah']} keluhan")
fig, ax = plt.subplots(1, 3, figsize=(20, 6))
sns.countplot(data=df_clean, x='Adaptasi', ax=ax[0], palette='Blues', order=['Sangat lambat', 'Lambat', 'Cepat', 'Sangat cepat'])
ax[0].set_title('Kecepatan Adaptasi Ritme PKU', fontweight='bold')
ax[0].set_xlabel('Tingkat Kecepatan')
ax[0].set_ylabel('Jumlah Mahasiswa')
ax[0].bar_label(ax[0].containers[0])
sns.countplot(data=df_clean, x='Kondisi_Akademik', ax=ax[1], palette='Oranges', order=['Sangat kewalahan', 'Kurang percaya diri', 'Cukup percaya diri', 'Sangat percaya diri'])
ax[1].set_title('Tingkat Kepercayaan Diri Akademik', fontweight='bold')
ax[1].set_xlabel('Kondisi')
ax[1].set_ylabel('Jumlah Mahasiswa')
ax[1].tick_params(axis='x', rotation=15)
ax[1].bar_label(ax[1].containers[0])
sns.barplot(data=distribusi_hambatan.head(8), x='Jumlah', y='Jenis Hambatan', ax=ax[2], palette='Reds_r')
ax[2].set_title('Top 8 Hambatan Performa Akademik', fontweight='bold')
ax[2].set_xlabel('Jumlah Keluhan')
ax[2].set_ylabel('')
ax[2].bar_label(ax[2].containers[0], padding=3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
distribusi_minat = df_clean['Minat_Bantuan'].value_counts().reset_index()
distribusi_minat.columns = ['Minat', 'Jumlah']
program_split = df_clean['Program_Dibutuhkan'].str.split(',').explode().str.strip()
distribusi_program = program_split.value_counts().reset_index()
distribusi_program.columns = ['Jenis Program', 'Jumlah']
matkul_split = df_clean['Matkul_Sulit'].str.lower().str.split(',').explode().str.strip()
matkul_split = matkul_split.str.title()
distribusi_matkul = matkul_split.value_counts().reset_index()
distribusi_matkul.columns = ['Mata Kuliah', 'Jumlah']
fig, ax = plt.subplots(1, 3, figsize=(24, 7))
sns.barplot(data=distribusi_minat, x='Minat', y='Jumlah', hue='Minat', legend=False, ax=ax[0], palette='Greens_r', order=['Sangat tertarik', 'Tertarik', 'Tidak tertarik', 'Sangat tidak tertarik'])
ax[0].set_title('Minat KM PKU Terhadap Program Bantuan', fontsize=14, fontweight='bold')
ax[0].set_xlabel('Tingkat Ketertarikan', fontsize=12)
ax[0].set_ylabel('Jumlah Mahasiswa', fontsize=12)
ax[0].tick_params(axis='x', rotation=15)
ax[0].bar_label(ax[0].containers[0])
sns.barplot(data=distribusi_program.head(6), x='Jumlah', y='Jenis Program', hue='Jenis Program', legend=False, ax=ax[1], palette='mako_r')
ax[1].set_title('Top 6 Program Bantuan Paling Diinginkan', fontsize=14, fontweight='bold')
ax[1].set_xlabel('Jumlah Permintaan', fontsize=12)
ax[1].set_ylabel('')
ax[1].bar_label(ax[1].containers[0], padding=3)
sns.barplot(data=distribusi_matkul.head(10), x='Jumlah', y='Mata Kuliah', hue='Mata Kuliah', legend=False, ax=ax[2], palette='Purples_r')
ax[2].set_title('Top 10 Matkul Paling Butuh Bantuan', fontsize=14, fontweight='bold')
ax[2].set_xlabel('Jumlah Keluhan', fontsize=12)
ax[2].set_ylabel('')
ax[2].bar_label(ax[2].containers[0], padding=3)
plt.tight_layout()
plt.show()
```

```python
print("REKAP MINAT BANTUAN AKADEMIK")
for index, row in distribusi_minat.iterrows():
    print(f"- {row['Minat']} : {row['Jumlah']} mahasiswa")
print("\nREKAP PROGRAM BANTUAN PALING DIBUTUHKAN")
for index, row in distribusi_program.iterrows():
    print(f"- {row['Jenis Program']} : {row['Jumlah']} permintaan")
print("\nTOP 10 MATKUL PALING BIKIN PUSING")
for index, row in distribusi_matkul.head(10).iterrows():
    print(f"- {row['Mata Kuliah']} : {row['Jumlah']} keluhan")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
distribusi_waktu = df_clean['Waktu_Ideal'].value_counts()
distribusi_kehadiran = df_clean['Kehadiran_Akpres'].value_counts().reset_index()
distribusi_kehadiran.columns = ['Kehadiran', 'Jumlah']
lomba_split = df_clean['Bantuan_Lomba'].str.split(',').explode().str.strip()
distribusi_lomba = lomba_split.value_counts().reset_index()
distribusi_lomba.columns = ['Jenis Bantuan Lomba', 'Jumlah']
print("REKAP WAKTU IDEAL PROGRAM")
print(distribusi_waktu)
print("\nREKAP RAPOR KEHADIRAN AKPRES")
for idx, row in distribusi_kehadiran.iterrows():
    print(f"- {row['Kehadiran']} : {row['Jumlah']} mahasiswa")
print("\nREKAP BANTUAN LOMBA PALING DIBUTUHKAN")
for idx, row in distribusi_lomba.iterrows():
    print(f"- {row['Jenis Bantuan Lomba']} : {row['Jumlah']} permintaan")
fig, ax = plt.subplots(1, 3, figsize=(24, 8))
ax[0].pie(distribusi_waktu, labels=distribusi_waktu.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
ax[0].set_title('Waktu Ideal Program Bantuan Akademik', fontsize=14, fontweight='bold')
import textwrap
labels_wrapped = [textwrap.fill(label, 15) for label in distribusi_kehadiran['Kehadiran']]
sns.barplot(data=distribusi_kehadiran, x='Kehadiran', y='Jumlah', hue='Kehadiran', legend=False, ax=ax[1], palette='flare')
ax[1].set_title('Sejauh Mana Akpres Hadir sbg "Rumah Kedua"?', fontsize=14, fontweight='bold')
ax[1].set_xticklabels(labels_wrapped)
ax[1].set_xlabel('')
ax[1].set_ylabel('Jumlah Mahasiswa', fontsize=12)
ax[1].bar_label(ax[1].containers[0], padding=3)
sns.barplot(data=distribusi_lomba.head(5), x='Jumlah', y='Jenis Bantuan Lomba', hue='Jenis Bantuan Lomba', legend=False, ax=ax[2], palette='crest')
ax[2].set_title('Top 5 Bantuan Lomba Paling Dibutuhkan', fontsize=14, fontweight='bold')
ax[2].set_xlabel('Jumlah Permintaan', fontsize=12)
ax[2].set_ylabel('')
ax[2].bar_label(ax[2].containers[0], padding=3)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
semua_harapan = ' '.join(df_clean['Harapan_Ormawa'].astype(str).tolist()).lower()
kata_buang = set(['dan', 'yang', 'di', 'ke', 'dari', 'untuk', 'pada', 'dengan', 'itu', 'ini', 'adalah', 'bisa', 'ada', 
                  'saya', 'kami', 'kita', 'mereka', 'juga', 'atau', 'lebih', 'agar', 'kalau', 'buat', 'biar', 'kalau',
                  'sih', 'dong', 'ya', 'banget', 'terus', 'sama', 'kayak', 'kalo', 'yg', 'aja', 'nya', 'pku', 'ormawa',
                  'eksekutif', 'akpres', 'bantuan', 'akademik', 'program', 'seperti', 'dalam', 'hal'])
wordcloud = WordCloud(width=1000, height=500, 
                      background_color='white', 
                      stopwords=kata_buang, 
                      colormap='inferno',
                      max_words=100).generate(semua_harapan)
plt.figure(figsize=(15, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Awan Kata (Wordcloud) Harapan & Saran KM PKU', fontsize=20, fontweight='bold')
plt.show()
```
